```python
import torch
import numpy
import numpy as np
```

```python
test_errors_2 = torch.load('recon_2.pt')
test_errors_16 = torch.load('recon_16.pt')
```

```python
import matplotlib.pyplot as plt
import math
plt.rcParams['font.size'] = 20

x = [2000,4000,6000,8000,10000]
y = np.mean(test_errors_2, 1)
y_error = np.std(test_errors_2, 1)
plt.errorbar(x, y ,yerr = y_error, marker = 'o', label = 'd_int = 2', capsize=3, ecolor='tab:orange')

y = np.mean(test_errors_16, 1)
y_error = np.std(test_errors_16, 1)
plt.errorbar(x, y,yerr = y_error, marker = 'o', label = 'd_int = 16',
             capsize=3, color = 'tab:green', ecolor='tab:red')
plt.xlabel("# Datapoints")
plt.ylabel("Reconstruction Error")

plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.2),ncol=2)
plt.savefig('GAN_recon.pdf', bbox_inches='tight')
```
